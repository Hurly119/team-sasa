```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time

```

```python
from time import sleep
from selenium.webdriver.common.action_chains import ActionChains

PATH = r"C:\Users\justine.padayao\Documents\chromedriver.exe"
driver = webdriver.Chrome(PATH)


links = []

for i in range(1,10):
  driver.get(f"https://www.pna.gov.ph/articles/search?q=vaccine&c=&p={i}")
  print(f"This is page number {i}")
  elems = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CSS_SELECTOR,".media-heading [href]")))
  elems = driver.find_elements_by_css_selector(".media-heading [href]")
  for elem in elems:
    link = elem.get_attribute('href')
    if link not in links:
      print(link)
      links.append(link)
  sleep(25)

driver.quit()

```

```python
PATH = r"C:\Users\justine.padayao\Documents\chromedriver.exe"
driver = webdriver.Chrome(PATH)
# driver.get("https://cnnphilippines.com/search/?q=covid+vaccine")

article = []

for link in links:
  try:
    driver.get(link)
    contents = driver.find_element_by_class_name("page-content").text
    title = driver.find_element_by_class_name("page-header").text
    authors = driver.find_element_by_class_name("author").text.replace("By ",'')
    published_date = driver.find_element_by_class_name("date").text
    news_org = 'Philippines News Agency'
    print(contents)
    print(title)
    print(authors)
    print(published_date)
    row_record = {'url': link, 'titles': title, 'authors': authors, 'published_date': published_date, 'articles_content': contents,'news_org': news_org} 
    article.append(row_record)
    sleep(30)
# print(news_org)
  except Exception as e:
    print(f'SCRAPING ERROR: {e}')
    pass
  continue

# import pandas as pd 
# df = pd.DataFrame(article)
# df.info()
# df.to_csv('Philippine News Agency Scraped Articles.csv')
```
