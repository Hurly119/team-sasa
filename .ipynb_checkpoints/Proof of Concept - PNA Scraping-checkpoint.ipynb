```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time

```

```python
# Links for Philippine News Agency

PATH = r"C:\Users\justine.padayao\Documents\chromedriver.exe"
driver = webdriver.Chrome(PATH)

driver.get("https://www.pna.gov.ph/articles/search?q=covid+vaccine")

elems = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CSS_SELECTOR,".media-heading [href]")))
elems = driver.find_elements_by_css_selector(".media-heading [href]")
links = [elem.get_attribute('href') for elem in elems]
print(links)

driver.quit()
```

## Pagination

```python
from time import sleep

PATH = r"C:\Users\justine.padayao\Documents\chromedriver.exe"
driver = webdriver.Chrome(PATH)
driver.get("https://cnnphilippines.com/search/?q=covid+vaccine")


headlines = []
links = []

for i in range(2,6):

  elems = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CSS_SELECTOR,".col-sm-8 [href]")))
  elems = driver.find_elements_by_css_selector(".col-sm-8 [href]")
  for elem in elems:
    link = elem.get_attribute('href')
    links.append(link)
    
  elems = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CSS_SELECTOR,".col-sm-8 [href]")))
  elems = driver.find_elements_by_css_selector(".col-sm-8 [href]")
  for elem in elems:
    headline = elem.text
    headlines.append(headline)
    
  i = str(i)
  element = driver.find_element_by_xpath(f"/html/body/div[2]/div/section[1]/section/div/div/div[1]/section/div/div/div/div[2]/ul/li[{i}]/span")
  element.click()
  sleep(25)

driver.quit()

print(headlines)
```

```python
PATH = r"C:\Users\justine.padayao\Documents\chromedriver.exe"
driver = webdriver.Chrome(PATH)
driver.get("https://cnnphilippines.com/search/?q=covid+vaccine")

article = []

try:
  for link in links:
    driver.get(link)
    contents = driver.find_elements_by_css_selector("p")
    for content in contents:
      print(content.text)
      article.append(content.text)
    sleep(30)
except:
  print('There is no article scraped')
  pass

driver.quit()
```

```python
import pandas as pd
df = pd.DataFrame(list(zip(headlines,links,article)),columns = ['Headline','Links','Content'])
df = df.drop_duplicates(subset = 'Headline', keep = False)
df
```

```python
df.to_csv('CNN Web Scraping.csv')
```
