```python
nlp = spacy.load('en_core_web_lg')
stop_words = stopwords.words("english")
#lowercase,#no nos.#no punctuation #no stopwords
```

```python
for stopword in stop_words:
    lexeme = nlp.vocab[stopword]
    lexeme.is_stop = True
```

```python
def clean(text):
    '''
    This function takes in texts and converts it to unicode format so we can process it in Python.
    '''
    return str(''.join([i if ord(i) < 128 else ' ' for i in text]))

def removeOutlier(df,columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        df = (df[(df[column] >= Q1 - 1.5*IQR) & 
                (df[column] <= Q3 + 1.5*IQR)])
    return df

def removeStopWords(string):
    doc = nlp(clean(string))
    tokens = []
    for w in doc:
        if not w.is_stop:
            tokens.append(w.text)
    return " ".join(tokens)

def removePunctuation(string):
    doc = nlp(clean(string))
    tokens = []
    for w in doc:
        if not w.is_punct:
            tokens.append(w.text)
    return " ".join(tokens)

def removeNumbers(string):
    doc = nlp(clean(string))
    tokens = []
    for w in doc:
        if not w.like_num:
            tokens.append(w.text)
    return " ".join(tokens)

def cleanDoc(string):
    doc = nlp(clean(string))
    tokens = []
    for w in doc:
        if w.text != '\n' and not w.is_stop and not w.is_punct and not w.like_num:
            tokens.append(w.lemma_)
    return " ".join(tokens)

def VADER(string):
    sid = SentimentIntensityAnalyzer()
    return sid.polarity_scores(string)["compound"]
```

```python
philstar_df = pd.read_csv("./scraped_successful/philstar_scraped2.csv")
manila_df = pd.read_csv("./scraped_successful/manila_scraped.csv")
philstar_df["news_agency"] = "Philstar"
manila_df["news_agency"] = "Manila Times"
manila_df = manila_df.drop([column for column in manila_df.columns if "Unnamed" in column],axis=1)
philstar_df = philstar_df.drop([column for column in philstar_df.columns if "Unnamed" in column],axis=1)
```

```python
bow = [bow.split() for bow in df["clean_contents"]]
```

```python
bigram = gensim.models.Phrases(bow)
texts = [bigram[line] for line in bow]
```

```python
dictionary = Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
lsimodel = LsiModel(corpus=corpus, num_topics=10, id2word=dictionary)
hdpmodel = HdpModel(corpus=corpus, id2word=dictionary)
ldamodel = LdaModel(corpus=corpus, num_topics=10, id2word=dictionary)
```

```python
pyLDAvis.enable_notebook()
pyLDAvis.gensim.prepare(ldamodel,corpus,dictionary)
```

```python
pyLDAvis.enable_notebook()
pyLDAvis.gensim.prepare(hdpmodel,corpus,dictionary)
```

```python
lsitopics = [[word for word, prob in topic] for topicid, topic in lsimodel.show_topics(formatted=False)]
hdptopics = [[word for word, prob in topic] for topicid, topic in hdpmodel.show_topics(formatted=False)]
ldatopics = [[word for word, prob in topic] for topicid, topic in ldamodel.show_topics(formatted=False)]
```

```python
lsi_coherence = CoherenceModel(topics=lsitopics[:10],texts=texts,dictionary=dictionary,window_size=10).get_coherence()
hdp_coherence = CoherenceModel(topics=hdptopics[:10],texts=texts,dictionary=dictionary,window_size=10).get_coherence()
lda_coherence = CoherenceModel(topics=ldatopics[:10],texts=texts,dictionary=dictionary,window_size=10).get_coherence()
```

```python
def evaluate_bar_graph(coherences, indices):
    assert len(coherences) == len(indices)
    n = len(coherences)
    x = np.arange(n)
    plt.bar(x, coherences, width=0.2,tick_label=indices, align='center')
    plt.xlabel('Models')
    plt.ylabel('Coherence Value')
```

```python
evaluate_bar_graph([lsi_coherence, hdp_coherence, lda_coherence], ['LSI', 'HDP', 'LDA'])
```

```python
coherences = []
for num_topics in range(1,20):
    lsimodel = LsiModel(corpus=corpus, num_topics=num_topics, id2word=dictionary)
    lsitopics = [[word for word, prob in topic] for topicid, topic in lsimodel.show_topics(formatted=False)]
    lsi_coherence = CoherenceModel(topics=lsitopics[:10],texts=texts,dictionary=dictionary,window_size=10).get_coherence()
    coherences.append(lsi_coherence)
    

sns.lineplot(list(range(1,20)),coherences)
```
